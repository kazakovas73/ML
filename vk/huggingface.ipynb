```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-ru")

model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-ru")
```

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
newsgroups_train = fetch_20newsgroups(subset='train')["data"][:100]
```

```python
t = tokenizer(newsgroups_train['data'][0], return_tensors="pt").input_ids
m = model.generate(t)
decoded = tokenizer.decode(m[0], skip_special_tokens=True)
decoded
```

```python
from nltk import tokenize
import nltk
nltk.download('punkt')
```

```python
texts = []
for idx, val in enumerate(newsgroups_train):
    a = tokenize.sent_tokenize(val)[2:]
    for i in a:
        if len(texts) == 100:
            break
        texts.append(i)
    if len(texts) == 100:
            break
```

```python
for idx, text in enumerate(texts):
    t = tokenizer(text, return_tensors="pt").input_ids
    m = model.generate(t)
    decoded = tokenizer.decode(m[0], skip_special_tokens=True)
    
    print(f"{idx + 1} \nText: {text}\nTranslation: {decoded}")
```
